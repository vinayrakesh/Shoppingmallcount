```python
# Face Recognition

# Importing the libraries
import time
import cv2
count=0
male=0
female=0
# Loading the cascades
face_cascade = cv2.CascadeClassifier(r'Downloads\haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(r'Downloads\haarcascade_eye.xml')
from keras.models import load_model
from keras.preprocessing import image
import numpy as np
import cv2
model = load_model('face_detect.h5')
model.compile(loss='binary_crossentropy',
                      optimizer='adam',
                      metrics=['accuracy'])
from skimage.io import imread
from skimage.transform import resize
# Defining a function that will do the detections
def detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        print (faces.shape)
        print ("Number of faces detected: " + str(faces.shape[0]))
        print("Data Found")
        global count 
        time.sleep(0.2)
        count+=faces.shape[0]
        cv2.putText(frame, "Number of faces detected: " + str(faces.shape[0]), (10, 30),
					cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        img = resize(roi_color,(64,64))
        img = np.expand_dims(img,axis=0)
        if(np.max(img)>1):
            img = img/255.0
        prediction = model.predict_classes(img)
        print(prediction)
        FaceFileName = "face_" + str(y) + ".jpg"
        #cv2.imwrite(FaceFileName, roi_color)
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        cv2.putText(frame, "Prediction:"+str(prediction[0][0]) , (x-2, y-2),
					cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        global male,female
        if(prediction[0][0]==0):
            female=female+1
            print("Female : " + str(female))
            cv2.putText(frame, "Female : " + str(female) , (20, 50),
					cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
            
        else:
            male=male+1
            print("Male : " + str(male))
            cv2.putText(frame, "Male : " + str(male) , (100, 100),
					cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        print("count : " + str(count))
    return frame
# Doing some Face Recognition with the webcam
video_capture = cv2.VideoCapture(0)
while True:
    _,frame = video_capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas = detect(gray, frame)
    cv2.imshow('Video', canvas)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Total count : " + str(count))
        print("Total Male count : " + str(male))
        print("Total Female count : " + str(female))
        break
video_capture.release()
cv2.destroyAllWindows()
```
